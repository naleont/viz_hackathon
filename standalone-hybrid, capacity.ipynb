```python
import pandas as pd

df = pd.read_csv('C:/users/katak/PycharmProjects/BA2/capacity.csv')
df
```

```python
What percentage of energy production came from each type during the years 2014-2022?
```

```python
import numpy as np
import matplotlib.pyplot as plt

df.fillna(0)

df['standalone_total']=df['standalone_prior']+df['standalone_new']
df['hybrid_total']=df['hybrid_prior']+df['hybrid_new']
df
```

```python
N = 7
ind = np.arange(N) 
width = 0.1
  
sovals = df[df["type"] == "Solar"]["total_gw"]
bar1 = plt.bar(ind, sovals, width, color = 'yellow')
  
wvals = df[df["type"] == "Wind"]["total_gw"]
bar2 = plt.bar(ind+width, wvals, width, color='blue')
  
stvals = df[df["type"] == "Storage"]["total_gw"]
bar3 = plt.bar(ind+width*2, stvals, width, color = 'grey')

gvals = df[df["type"] == "Gas"]["total_gw"]
bar4 = plt.bar(ind+width*3, gvals, width, color = 'green')

nvals = df[df["type"] == "Nuclear"]["total_gw"]
bar5 = plt.bar(ind+width*4, nvals, width, color = 'pink')

cvals = df[df["type"] == "Coal"]["total_gw"]
bar6 = plt.bar(ind+width*5, cvals, width, color = 'brown')

ovals = df[df["type"] == "Other"]["total_gw"]
bar7 = plt.bar(ind+width*6, ovals, width, color = 'red')
  
plt.xlabel("Years")
plt.ylabel('GW')
plt.title("Distribution across different types of energy")
  
plt.xticks(ind+width,['2014', '2015', '2016', '2017', '2018', '2019', '2020'])
plt.legend( (bar1, bar2, bar3, bar4, bar5, bar6, bar7), ('Solar', 'Wind', 'Storage', 'Gas', 'Nuclear', 'Coal', 'Other') )

plt.savefig('distr.pdf', dpi=600)

plt.savefig('distr.png', dpi=600)
```

```python
x = df[df["type"] == "Solar"]["year"]
y1 = df[df["type"] == "Solar"]["standalone_total"]
y2 = df[df["type"] == "Solar"]["hybrid_total"]
  
plt.xlabel("Years")
plt.ylabel('GW')
plt.title("Standalone/hybrid distrubution for solar energy")

plt.bar(x, y1, color='pink')
plt.bar(x, y2, bottom=y1, color='green')
plt.legend(["Standalone", "Hybrid"])

plt.savefig('solar_hybrid.pdf', dpi=600)

plt.savefig('solar_hybrid.png', dpi=600)
```

```python
x = df[df["type"] == "Wind"]["year"]
y1 = df[df["type"] == "Wind"]["standalone_total"]
y2 = df[df["type"] == "Wind"]["hybrid_total"]
  
plt.xlabel("Years")
plt.ylabel('GW')
plt.title("Standalone/hybrid distrubution for wind energy")

plt.bar(x, y1, color='pink')
plt.bar(x, y2, bottom=y1, color='green')
plt.legend(["Standalone", "Hybrid"])
plt.savefig('wind_hybrid.pdf', dpi=600)

plt.savefig('wind_hybrid.png', dpi=600)
```
